## Agregar columna con el nombre del estado

En esta sección vamos a concatenar las reviews por estado, para ello para cada uno de los 52 archivos vamos a agregar una columna con el nombre del estado, luego de esto concatenaremos los archivos en uno solo.

```python
import os
import pandas as pd

# Directorio que contiene los archivos CSV filtrados
directorio_filtrados = 'csv/estados-filter'

# Recorrer los archivos dentro de la carpeta filtrados
for filename in os.listdir(directorio_filtrados):
    if filename.endswith(".csv"):
        filepath = os.path.join(directorio_filtrados, filename)
        print(filepath)
        nombre=filepath[26: -13]
        
        # Leer el archivo CSV filtrado
        estado_dataframe_filtrado = pd.read_csv(filepath)

        # Agregar una columna con el nombre del estado al DataFrame
        estado_dataframe_filtrado['nombre_estado'] = nombre

        # Construir la ruta de salida para guardar el DataFrame modificado
        ruta_salida = os.path.join('.', 'estados', f'{nombre}.csv')

        # Guardar el DataFrame modificado en un nuevo archivo CSV
        estado_dataframe_filtrado.to_csv(ruta_salida, index=False)
```

## Creacion del archivo concatenado

En esta seccion vamos a unificar cada uno de los archivos en un uno solo, cada archivo tendra una columna identificadora con el nombre del estado, finalmente almacenaremos este archivo filtrado en formato csv.gz

```python
import os
import pandas as pd

# Directorio que contiene los archivos CSV filtrados
directorio_filtrados = './estados'

# Lista para almacenar los DataFrames
dataframes = []

# Recorrer los archivos dentro de la carpeta filtrados
for filename in os.listdir(directorio_filtrados):
    if filename.endswith(".csv"):
        filepath = os.path.join(directorio_filtrados, filename)

        # Leer el archivo CSV
        df = pd.read_csv(filepath)

        # Agregar el DataFrame a la lista
        dataframes.append(df)

# Concatenar todos los DataFrames en uno solo
df_final = pd.concat(dataframes, ignore_index=True)

# Guardar el DataFrame final en un nuevo archivo CSV
estados_concatenados_path = os.path.join('../../','data','google','estados_concatenados.csv.gz')
df_final.to_csv(estados_concatenados_path, index=False,compression='gzip')
```
