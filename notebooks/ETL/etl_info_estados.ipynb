```python
import pandas as pd
```

```python
ruta_archivo = '../../data/google/info_estados_raw.csv.gz'

info = pd.read_csv(ruta_archivo)
```

```python
info
```

```python
nuevos_nombres = {
    'nombre_estado': 'Estado',
    ' poblacion': 'Poblacion',
    'PBI_pc': 'PBI',
    'area_kms': 'Area_Kms',
    'ingreso_medio': 'Ingreso_Medio',
}

info.rename(columns=nuevos_nombres, inplace=True)
```

```python
info
```

```python
info['Poblacion'] = info['Poblacion'].str.replace('.','').astype(int)
```

```python
info['Estado'].replace({'District_of_Columbia':'District of Columbia'},inplace=True)
```

```python
info['Poblacion']
```

```python
info.to_csv('../../data/google/info_estados_clear.csv.gz',compression='gzip',index=False)
```
