```python
import pandas as pd 
import os
import json
import pandas as pd
import pyarrow as pa
import pyarrow.parquet as pq
import gzip
```

```python
path_reviews = os.path.join('..','..','data','yelp','reviews.csv.gz')
reviews = pd.read_csv (path_reviews, compression='gzip')
```

# Extracci√≥n

```python
users = pd.read_parquet('./Datasets/Yelp/user-001.parquet')
```

```python
users
```

## Transformaci√≥n

```python
users = users[users['user_id'].isin(reviews['user_id'])]
```

### Divisi√≥n de datos para carga incremental

```python
users['date_create_user'] = pd.to_datetime(users['yelping_since']).dt.year

users_2022 = users[users['date_create_user'] == 2022]
users = users[users['date_create_user'] < 2022]
```

## Carga

```python
user_path = os.path.join('..','..','data','yelp','users.csv.gz')
users.to_csv (user_path, compression='gzip', index=False)

user_path_2022 = os.path.join('..','..','data','yelp','users_2022.csv.gz')
users.to_csv (user_path_2022, compression='gzip', index=False)
```
